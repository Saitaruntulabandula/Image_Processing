# Reading an Image File - Using OpenCV

```python
import cv2                                       #OpenCV is an Machine Learning Package for analyzing image patterns in python.
```

OpenCV stores colored images in BGR(Blue Green and Red) format.

# Gray Image

cv2.IMREAD_GRAYSCALE: It specifies to load an image in grayscale mode. Alternatively, we can pass integer value 0 for this flag.

```python
cv_img = cv2.imread("Image.jpg",0)               #Using OpenCV #Grey Image 
```

```python
#cv_img = cv2.imread("Image.jpg",cv2.IMREAD_GRAYSCALE) 
```

```python
cv2.imshow("Grey Image",cv_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(cv_img.shape)
```

# Colour Image

cv2.IMREAD_COLOR: It specifies to load a color image. Any transparency of image will be neglected. It is the default flag. Alternatively, we can pass integer value 1 for this flag.

```python
cv_img = cv2.imread("Image.jpg",1)               #Using OpenCV #Colour Image #cv2 uses BGR color format
```

```python
#cv_img = cv2.imread("Image.jpg",cv2.IMREAD_COLOR) 
```

```python
cv2.imshow("Colour Image",cv_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(cv_img.shape)
```

# Displaying images with different Colorscapes

# BGR Colorscape

BGR color space: OpenCVs default color space is RGB. However, it actually stores color in the BGR format. It is an additive color model where the different intensities of Blue, Green and Red

```python
img=cv2.imread("Image.jpg",1)
```

```python
B, G, R = cv2.split(img)
```

```python
cv2.imshow("RGB Image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow("blue", B)
cv2.waitKey(0)
 
cv2.imshow("Green", G)
cv2.waitKey(0)
 
cv2.imshow("red", R)
cv2.waitKey(0)
 
cv2.destroyAllWindows()
```

```python
print(img.shape) 
```

```python
print(img.dtype)
```

# HSV(Hue Saturation Value) Colorspace

HSV color space: It stores color information in a cylindrical representation of RGB color points. It attempts to depict the colors as perceived by the human eye. Hue value varies from 0-179, Saturation value varies from 0-255 and Value value varies from 0-255. It is mostly used for color segmentation purpose.

```python
hsv_img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
```

```python
H, S, V = cv2.split(hsv_img)
```

```python
cv2.imshow("HSV Image",hsv_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow("Hue", H)
cv2.waitKey(0)
 
cv2.imshow("Saturation", S)
cv2.waitKey(0)
 
cv2.imshow("Value", V)
cv2.waitKey(0)
 
cv2.destroyAllWindows()
```
